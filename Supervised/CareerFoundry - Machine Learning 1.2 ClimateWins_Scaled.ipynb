# 1. Importing data and libriaries

```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
# Defining the path
path = 'C:/Users/andyc/Machine Learning - ClimateWins'
```

```python
#importing the Climatewins data set
ClimateWins = pd.read_csv(os.path.join(path, 'Data Sets', 'Dataset-weather-prediction-dataset-processed.csv'), index_col = False)
```

```python
# Checking the first 5 rows of the ClimateWins data set
ClimateWins.head()
```

```python
# Reset index
# Renaming the index column to id
ClimateWins.reset_index(inplace=True)
ClimateWins.rename(columns={'index':'id'}, inplace=True)
ClimateWins.head()
```

```python
# Checking ClimateWins' info
ClimateWins.info()
```

# 2. Scaling

```python
# Checking rows and columns
ClimateWins.shape
```

```python
# Checking for missing values
ClimateWins.isnull().sum()
```

```python
# eliminate those records that have NaN values
ClimateWins.dropna(inplace=True)
```

```python
# Checking rows and columns after dropping missing values
ClimateWins.shape
```

No missing values are present in the data set.

```python
# Checking for duplicates
ClimateWins_dupes = ClimateWins[ClimateWins.duplicated]
```

# 3. Scaling

```python
# create scaler object using StandardScaler from sklearn.preprocessing
scaler = StandardScaler()
```

```python
# create a new dataframe with scaled data

ClimateWins_scaled = pd.DataFrame(scaler.fit_transform(ClimateWins), columns=ClimateWins.columns)
ClimateWins_scaled.head()
```

```python
# Overwriting the 'date' column in the scaled dataframe with the original 'date' column in our ClimateWins dataframe
ClimateWins_scaled['DATE'] = ClimateWins['DATE']
```

```python
ClimateWins_scaled.head()
```

```python
# Comparing the original data and the scaled data
ClimateWins.head()
```

```python
# exporting dataframe as pickle file
ClimateWins_scaled.to_csv(os.path.join(path, 'ClimateWins_scaled.csv'))
```
